```python
#we can see that dataset contains a lot of NaN which are not really usefull for use. We cant really extrapolate what
#value should be there by some distribution because when the whole row is NaN the measurement_id is -1 which holds
#no usefull information for training

df = df.dropna()
df.shape
df.head(10)
```

```python
import sklearn
import numpy as np

df_transformation = sklearn.pipeline.Pipeline([
                ("preprocess", sklearn.compose.ColumnTransformer([
                    #("onehot", sklearn.preprocessing.OneHotEncoder(handle_unknown="ignore"), ["MeasurementId"]),
                    ("categorical", sklearn.preprocessing.OrdinalEncoder(categories=[[False, True],[False, True],[False, True]]),
                     ["PumpFailed","SlowStart","SlowEnd"] ),
                ]))
            ])

transformed_array = df_transformation.fit_transform(df)

dense_array = transformed_array.toarray()


print(dense_array)
```
