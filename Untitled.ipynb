```python
#merging the two dataframes
df_merged = pd.merge(df, df_parquet, on=['MachineId', 'MeasurementId'], how='left')
df_merged.head()
```

```python
df_merged = df_merged.dropna()
df_merged.shape
```

```python
#we can see that dataset contains a lot of NaN which are not really usefull for use. We cant really extrapolate what
#value should be there by some distribution because when the whole row is NaN the measurement_id is -1 which holds
#no usefull information for training

pd.set_option('display.float_format', lambda x: '%.3f' % x)
df_merged.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(data=df_merged[:2_000_000], x='PumpFailed', y='Pressure')
plt.xlabel('Pump Failed')
plt.ylabel('Pressure')
plt.title('Pressure Distribution by Pump Failure Status')
plt.show()
```

```python
#from above plot we can see outliers in both cases when pump did or didnt fail so very high outliers doesnt imply by themselves
#that pump will fail
```

```python
df_merged["SlowStart"] = df_merged["SlowStart"].astype(int)
df_merged["SlowEnd"] = df_merged["SlowEnd"].astype(int)
df_merged["PumpFailed"] = df_merged["PumpFailed"].astype(int)

aggregation = df_merged.groupby(['MachineId']).agg({'Pressure': ['mean', 'std', 'min', 'max'],
                                                    'PumpFailed': ['min', 'max'],
                                                    'SlowStart': ['min', 'max'],
                                                    'SlowEnd': ['min', 'max']}).reset_index()

aggregation.set_index

plt.figure(figsize=(10, 8))
sns.heatmap(aggregation.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
#we can see some interesting correlations such that when slow start is true or slow end is true then there is tendency for pressure to be higher
#at peaks. Also we can see that when slow start is true then slow end has tendency to be true as well

#with respect to the actual thing we want to predict - pump failure, we can see that there is some correlation between pump failure, max pressure,
#slow start and a bit less with slow end
```

```python
df_merged[2_000_000:2_000_010]
```

```python
#as we observed taking only mean isnt enough because there are peaks in pressure in both cases when pump failed and when it didnt.
#so we will take min, max and mean and std

df_merged_temp = df_merged.groupby(['MachineId']).agg({'Pressure': ['mean', 'min', 'max', 'std']})
df_merged_temp.columns = ['_'.join(col) for col in df_merged_temp.columns.values]
df_merged_temp = df_merged_temp.reset_index()
df_final = pd.merge(df, df_merged_temp, on='MachineId', how='left')
df_final = df_final.dropna()
```

```python
df_final.head()
```

```python
df_final["SlowStart"] = df_final["SlowStart"].astype(int)
df_final["SlowEnd"] = df_final["SlowEnd"].astype(int)
df_final["PumpFailed"] = df_final["PumpFailed"].astype(int)
```

```python
#based on our finding we i chose to select random forest classifier for this problem. This classifier is great for table-like data.
#Also its great for cases when the data doesnt have linear relationship between them which is what we observed above

import sklearn
import numpy as np

from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder, RobustScaler, MinMaxScaler
from sklearn.impute import SimpleImputer

train_data, test_data, train_target, test_target = sklearn.model_selection.train_test_split(df_final.drop('PumpFailed', axis=1), df_final['PumpFailed'], test_size=0.2)

#drop machine id and measurement id because they are not usefull for training
train_data.drop(['MachineId', 'MeasurementId'], axis=1, inplace=True)

numeric_features = ['Pressure_mean', 'Pressure_min', 'Pressure_max', 'Pressure_std']
numeric_transformer = sklearn.pipeline.Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')), # replace missing values with the mean
    ('scaler', MinMaxScaler())]) # initialy i tought about standart scaler but minmax scaler gave better results

categorical_features = ['SlowStart', 'SlowEnd']
categorical_transformer = sklearn.pipeline.Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')), # replace missing values with the mean
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])    # one-hot encode the categorical features

model = [
            ("poly", sklearn.preprocessing.PolynomialFeatures(2)),
            ("rf", sklearn.ensemble.RandomForestClassifier(n_estimators=100, max_depth=30, verbose=1, max_features='log2')),
        ]

model = sklearn.pipeline.Pipeline([
                ("preprocess", sklearn.compose.ColumnTransformer([
                    ("numeric", numeric_transformer, numeric_features),
                    ("categorical", categorical_transformer, categorical_features),
                ]))
            ] + model)

model.fit(train_data, train_target)

acc = model.score(test_data, test_target)


# # Fit the pipeline to the training data
# model.fit(train_data, train_target)

# # Get the model's score on the test set
# acc = model.score(test_data, test_target)

# {'poly__degree': 2, 'rf__max_depth': 30, 'rf__max_features': 'log2', 'rf__n_estimators': 100} best result after grid search
# grid_search = GridSearchCV(model, param_grid={
#     "poly__degree": [1, 2, 3],
#     "rf__n_estimators": [10, 100, 1000],
#     "rf__max_depth": [10, 20, 30],
#     "rf__max_features": ["auto", "sqrt", "log2"],
# }, cv=5, verbose=2)

# grid_search.fit(train_data, train_target)

# print(grid_search.best_params_)

# # Get the model's score on the test set
# acc = grid_search.score(test_data, test_target)
```
